This notebook contains code to generate some of the visualizations presented in the lecture slides.

```python
import numpy as np
import torch
from torch import tensor
from torch.autograd import Variable
import matplotlib.pyplot as plt
```

## Activation functions



### Step (Heaviside) function

```python
x = torch.arange(-10, 10, 0.1)

f, ax = plt.subplots(1, 1, figsize=(7, 5))

ax.plot(x, torch.heaviside(x, tensor([0.5])), label='function', lw=2)
ax.plot(x, torch.zeros(x.shape), label='gradient')
ax.grid()
ax.legend()

plt.savefig('actfunc_step.png', dpi=300)
```

### Sigmoid function

```python
# instantiate Sigmoid function
sigmoid = torch.nn.Sigmoid()

f, ax = plt.subplots(1, 1, figsize=(7, 5))

# create Variable with a gradient
x = Variable(torch.arange(-10, 10, 0.1), requires_grad=True)

# compute y values and gradients
y = sigmoid(x)
y.backward(torch.ones_like(x))

# plot function values and gradients
ax.plot(x.detach().numpy(), sigmoid(x.detach()), lw=2, label='function')
ax.plot(x.detach().numpy(), x.grad, label='gradient')
ax.grid()
ax.legend()

plt.savefig('actfunc_sigmoid.png', dpi=300)
```

### Tanh function

```python
# instantiate tanh function
tanh = torch.nn.Tanh()

f, ax = plt.subplots(1, 1, figsize=(7, 5))

# create Variable with a gradient
x = Variable(torch.arange(-10, 10, 0.1), requires_grad=True)

# compute y values and gradients
y = tanh(x)
y.backward(torch.ones_like(x))

# plot function values and gradients
ax.plot(x.detach().numpy(), tanh(x.detach()), lw=2, label='function')
ax.plot(x.detach().numpy(), x.grad, label='gradient')
ax.grid()
ax.legend()

plt.savefig('actfunc_tanh.png', dpi=300)
```

### ReLU function

```python
# instantiate ReLU function
relu = torch.nn.ReLU()

f, ax = plt.subplots(1, 1, figsize=(7, 5))

# create Variable with a gradient
x = Variable(torch.arange(-10, 10, 0.1), requires_grad=True)

# compute y values and gradients
y = relu(x)
y.backward(torch.ones_like(x))

# plot function values and gradients
ax.plot(x.detach().numpy(), relu(x.detach()), lw=2, label='function')
ax.plot(x.detach().numpy(), x.grad, label='gradient')
ax.grid()
ax.legend()

plt.savefig('actfunc_relu.png', dpi=300)
```
